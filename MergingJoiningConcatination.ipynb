```python
#DATAFRAME -1
employees=pd.DataFrame({
    'Employee_id':[1,2,3,4,5],
    'Nama':['John','Lucky','Ronny','Max','Arnav'],
    'Department':['HR','IT','Finance','IT','HR']
})

#DATAFRAME-2
salaries=pd.DataFrame({
    'Employee_id':[1,2,3,6,7],
    'Salary':[90000,70000,65000,73000,90000],
    'Bonus':[15000,10000,8000,10000,15000]
})
```

```python
pd.merge(employees,salaries,on='Employee_id',how='inner')
```

```python
pd.merge(employees,salaries,on='Employee_id',how='outer')
```

```python
pd.merge(employees,salaries,on='Employee_id',how='left')
```

```python
pd.merge(employees,salaries,on='Employee_id',how='right')
```

```python
#CONCATINATION OF 2 FRAMES
```

```python
df1=pd.DataFrame({
    "A":['A0','A1','A2'],
    "B":['B0','B1','B2'],
    "C":['C0','C1','C2'],
})
df2=pd.DataFrame({
    "A":['A3','A4','A5'],
    "B":['B3','B4','B5'],
    "C":['C3','C4','C5'],
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
#JOINING 2 DATAFRAMES
```

```python
df1=pd.DataFrame({
    'Name':['John','Raj','Lucky']
},index=[1,2,3])

df2=pd.DataFrame({
    'score':[90,78,81]
},index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```
