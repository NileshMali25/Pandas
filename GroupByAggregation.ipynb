```python
import numpy as np
import pandas as pd
```

```python
#GROUPBY
```

```python
data={
    'Category':['A','B','A','B','A','B','A','B'],
    'Store':['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales':[100,150,200,230,140,300,250,120],
    'Quantity':[10,13,12,15,11,20,18,12],
    'Date':pd.date_range("2026-01-01",periods=8)
}
df=pd.DataFrame(data)
```

```python
df
```

```python
cat=df.groupby('Category')
for i,v in cat:
    print(i)
    print(v)
```

```python
#GROUPBY CATEGORY AND CALCULATE SUM OF SALES
cat=df.groupby('Category')['Sales'].sum()
cat
```

```python
#GROUPBY STORE AND CALCULATE SUM OF SALES
```

```python
cat=df.groupby('Store')['Sales'].sum()
cat
```

```python
#GROUPBY MULTIPLE COLUMN
#GROUPBY CATEGORY AND STORE
```

```python
cat=df.groupby(['Category','Store'])['Sales'].sum()
cat
```
