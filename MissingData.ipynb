```python
import numpy as np
import pandas as pd
```

```python
#FINDING MISSING DATA
```

```python
data={
    'A':[1,2,3,4,5],
    'B':[1,np.nan,3,4,5],
    'C':[1,2,np.nan,np.nan,5],
    'D':[1,np.nan,3,np.nan,np.nan]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
#REMOVING MISSING DATA
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3)
```

```python
#FILLING THE MISSING DATA
```

```python
df.fillna(0)
```

```python
values={'A':0,'B':11,'C':35,'D':49}
df.fillna(value=values)
```
